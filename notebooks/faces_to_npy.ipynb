This notebook converts a specific number of images from the `img_align_celeba` dataset to greyscale and saves them in a numpy array.  
This then could be used for a DCGAN that takes 1x32x32 images.  
Note that the images do have to be transformed within `numpy_dataset()` in the DCGAN notebook: 

    transform=transforms.Compose([
                               transforms.Resize(image_size),
                               transforms.CenterCrop(image_size),
                               transforms.ToTensor(),
                               transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)),
                           ])

```python
import numpy as np
import glob as glob
import matplotlib.pyplot as plt

from PIL import Image
from tqdm import tqdm

%matplotlib inline
%config InlineBackend.print_figure_kwargs={'facecolor' : "w"} # Make sure the axis background of plots is white, this is usefull for the black theme in JupyterLab
```

# Load images

```python
paths = glob.glob('F:\Jupyterlab\DCGAN\data\img_align_celeba - complete\*.jpg')
print('Detected %s images' % len(paths))
```

```python
"""
Loads images as PIL images from paths.
Saves the PIL images as a numpy array containing these PIL images.
"""

paths = paths[:10**2] # selected the first 100k images

images = []

for _ , path in enumerate(tqdm(paths)):
    image = Image.open(path) # opens PIL images
    #image = np.expand_dims(image, axis=0) # add channel dimension (N_images, C, H, W)
    images.append(image)
```

# Convert the list of images to an array and save the array

```python
%%time
images = np.array(images)
```
